```python
url = 'https://www.kinfa.or.kr/customerService/boardFAQ.do'

options = webdriver.ChromeOptions()	
#     options.add_argument('headless')	
options.add_argument('disable-gpu')	
driver = webdriver.Chrome('chromedriver.exe', options=options)
driver.get(url)
time.sleep(1)

total_qnaList = list()
fileName = '8_서민금융진흥원'
csv_file = open(str(fileName) + '.csv', "w", encoding = 'utf-8-sig', newline = '')

columns = ['키워드', '질문', '답변']

wr = csv.writer(csv_file)    
wr.writerow(columns)

page = 3
while page <= 12:
    qna = list()
    keyword_list = driver.find_elements_by_tag_name('#faq_con > ol > li > h3 > span')
    question_list = driver.find_elements_by_tag_name('#faq_con > ol > li > h3 > a')
    answer_list = list()

    for idx, question in enumerate(question_list):
        question.click()
        time.sleep(1)
        answer = driver.find_element_by_xpath('//*[@id="faq_con"]/ol/li[%s]/div' % str(idx+1))
        answer_list.append(answer.text.replace('\n',''))
    
    for keyword, question, answer in zip(keyword_list, question_list, answer_list):
        temp_qna = list()
        temp_qna.append(keyword.text)
        temp_qna.append(question.text)
        temp_qna.append(answer)
        qna.append(temp_qna)
    
    total_qnaList.append(qna)
    print(qna)

    try:
        next_page = driver.find_element_by_tag_name('#listForm > div.pagination > a:nth-of-type(%s)' % str(page))
        next_page.click()
        print('====================', page-2, '쪽 수집 완료===========================')
        page += 1
        if page == 13:
            page = 3
    except:
        print('마지막 페이지')
        break


for i in range(len(total_qnaList)):
    for j in range(len(total_qnaList[i])):
        wr.writerow(total_qnaList[i][j])

print('==================== 저장 완료 =======================')
csv_file.close()
```

```python
def get_qnaContent(url):
    qna = list()

    options = webdriver.ChromeOptions()	
    #     options.add_argument('headless')	
    options.add_argument('disable-gpu')	
    driver = webdriver.Chrome('chromedriver.exe', options=options)
    driver.get(url)
    time.sleep(1)

    keyword_list = driver.find_elements_by_tag_name('#faq_con > ol > li > h3 > span')
    question_list = driver.find_elements_by_tag_name('#faq_con > ol > li > h3 > a')
    answer_list = list()

    for idx, question in enumerate(question_list):
        question.click()
        time.sleep(1)
        answer = driver.find_element_by_xpath('//*[@id="faq_con"]/ol/li[%s]/div' % str(idx+1))
        answer_list.append(answer.text.replace('\n',''))
    
    for keyword, question, answer in zip(keyword_list, question_list, answer_list):
        temp_qna = list()
        temp_qna.append(keyword.text)
        temp_qna.append(question.text)
        temp_qna.append(answer)
        qna.append(temp_qna)
    return qna
```
