```python
url = 'http://finlife.fss.or.kr/board/selectArticleList.do?bltbId=BM000000000000000003&menuId=2000112'
options = webdriver.ChromeOptions()	
#     options.add_argument('headless')	
options.add_argument('disable-gpu')	
driver = webdriver.Chrome('chromedriver.exe', options=options)
driver.get(url)
time.sleep(1)


total_info_list = list()

for idx in range(2, 6):
    try:
        #페이지 수집
        for i in range(1, 11):
            info_dict = dict()
            info_dict['번호'] = driver.find_element_by_tag_name('tr:nth-of-type(%s) > td:nth-of-type(1)' % (str(i))).text
            detail_page = driver.find_element_by_tag_name('tr:nth-of-type(%s) > td.defaul-text > a' %(str(i)))
            detail_page.click()
            time.sleep(1)


            # 정보 수집
            info_dict['제목'] = driver.find_element_by_tag_name('tr:nth-of-type(1) > td > strong').text
            info_dict['작성자'] = driver.find_element_by_tag_name('tr.admin-Write-peo > td.txt_left').text
            info_dict['등록일'] = driver.find_element_by_tag_name("tr.admin-Write-peo > td:nth-of-type(2)").text
            info_dict['조회수'] = driver.find_element_by_tag_name('td:nth-of-type(3)').text
            info_dict['답변'] = driver.find_element_by_tag_name("tbody > tr:nth-of-type(3) > td").text

            driver.back()
            time.sleep(1)
            total_info_list.append(info_dict)
            print(info_dict)

            print('==============================')
        
        try:
            next_page = driver.find_element_by_tag_name('#contents > div.sub_content_area > div > div > a:nth-of-type(%s)' % str(idx))
            next_page.click()
        except:
            pass
        
    
    except:
        print('마지막 페이지')
        break
```

```python

fileName = '6_금융감독원파인(금융상품한눈에)'
csv_file = open(str(fileName) + '.csv', "w", encoding = 'utf-8-sig', newline = '')

columns = ['번호', '제목', '작성자', '등록일', '조회수', '답변']

wr = csv.writer(csv_file)    
wr.writerow(columns)

for qna in total_info_list:
    wr.writerow([qna.get(col, 'NULL') for col in columns])

csv_file.close()
```
