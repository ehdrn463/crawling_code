```python
def get_qnaList(url):

    req = requests.get(url)
    cont = req.content
    soup = BeautifulSoup(cont, 'html.parser')

    qna_list = soup.select('td.t_al > a')

    for idx, qna in enumerate(qna_list):
        qna_list[idx] = 'http://phishing-keeper.fss.or.kr/fss/vstop/bbs/' + qna['href']

    return qna_list
```

```python
def get_qnaContent(url):
    req = requests.get(url)
    cont = req.content
    soup = BeautifulSoup(cont, 'html.parser')
    
    qna = dict()
    qna['제목'] = soup.select_one('tbody > tr:nth-of-type(1) > td').text
    qna['등록일'] = soup.select('tbody > tr:nth-of-type(2) > td')[0].text
    qna['조회수'] = soup.select('tbody > tr:nth-of-type(2) > td')[1].text
    
    qna['담당부서'] = soup.select('tbody > tr:nth-of-type(3) > td')[0].text
    qna['담당팀'] = soup.select('tbody > tr:nth-of-type(3) > td')[1].text
    qna['문의'] = soup.select('tbody > tr:nth-of-type(3) > td')[2].text
    try:
        qna['첨부파일'] = soup.select_one('tbody > tr:nth-of-type(4) > td > a')['href']
    except:
        qna['첨부파일'] = 'NULL'
    qna['답변'] = soup.select_one('tbody > tr:nth-of-type(5) > td').text

    
    
    return qna
```

```python
def main():
    total_qnaList = list()
    fileName = '금융감독원(보이스피싱지킴이)'
    csv_file = open(str(fileName) + '.csv', "w", encoding = 'utf-8-sig', newline = '')
    
    columns = ['제목', '등록일', '조회수', '담당부서', '담당팀', '문의', '첨부파일', '답변']
    
    wr = csv.writer(csv_file)    
    wr.writerow(columns)
    
    for idx in range(1, 4):
        url = 'http://phishing-keeper.fss.or.kr/fss/vstop/bbs/list.jsp?url=/fss/vstop/1353319778171&bbsid=1353319778171&page=%s'%(str(idx))
        qnaList = get_qnaList(url)
        
        for qna in qnaList:
            qnaContent = get_qnaContent(qna)
            print(qnaContent)
            total_qnaList.append(qnaContent)
    
    
    # '모든 수집 복지서비스'에 대한 반복문 -> 복지 서비스 하나
    for info_dict in total_qnaList:
        # 'columns'을 돌면서 info_dict가 해당 col을 가지고 있으면 정보를 적고, 없으면 NULL
        wr.writerow([info_dict.get(col, 'NULL') for col in columns])
    
    print('==================== 저장 완료 =======================')
    csv_file.close()
             
    
```

```python
if __name__ == "__main__":
    main()
```
