```python
def clickNcrawl(url):
    """
    (string) -> None
        Args:
            url(string): Q&A 리스트
    Return:
        '질문'을 눌러야 답변내용이 나옴.
        '질문' 모두 클릭해서 답변 수집할 수 있도록 해주는 과정
        None
    """
    options = webdriver.ChromeOptions()	
    options.add_argument('headless')	
    options.add_argument('disable-gpu')	
    driver = webdriver.Chrome('chromedriver.exe', options=options)
    driver.get(url)
    
    question_list = driver.find_elements_by_tag_name('#contents > div.faqList3 > dl > dt')
    
    for question in question_list:
        question.click()
        time.sleep(1)
        
    question_list = driver.find_elements_by_tag_name('#contents > div.faqList3 > dl > dt')
    answer_list = driver.find_elements_by_tag_name('#contents > div.faqList3 > dl > dd')
    
    total_list = list()
    for idx in range(len(question_list)):
        total_list.append([question_list[idx].text, answer_list[idx].text])
    return total_list
        
    
```

```python
def main():
    current_path = os.getcwd()
    data_path = os.getcwd() + '\\복지로2_자주묻는질문\\'
    
#     fileName = data_path + '복지로2_자주묻는질문'
    fileName = current_path + '\\복지로2_자주묻는질문'
    csv_file = open(str(fileName) + '.csv', "w", encoding = 'utf-8-sig', newline = '')
    wr = csv.writer(csv_file)
    wr.writerow(['질문', '답변'])
    
    for page in list(range(1, 11)):
        url = 'http://online.bokjiro.go.kr/apl/faq/aplFaq.do?srchTabCode=006&srchTabCatId=APL092&srchSelCode=&srchSelCatId=&srchType=SRCH&srchSelCodeCatId=&srchKeyword=&pageUnit=10&pageIndex=%d'%(page)
        total_list = clickNcrawl('http://online.bokjiro.go.kr/apl/faq/aplFaq.do')
        
        for total in total_list:
            print(total)
            wr.writerow(total)
        print('====================================')
        print(page, '수집 완료')
        print('====================================')
            
    csv_file.close()
    print('복지로2 자주묻는 질문 파일 저장 완료')
            

        
```
