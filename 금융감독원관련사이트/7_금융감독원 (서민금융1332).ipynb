```python
def get_qnaList(url):

    req = requests.get(url, timeout = 30)
    cont = req.content
    soup = BeautifulSoup(cont, 'html.parser')

    qna_list = soup.select('td.bbs_subj > a')

    for idx, qna in enumerate(qna_list):
        qna_list[idx] = 'http://www.fss.or.kr/s1332/etc/' + qna['href']

    return qna_list
```

```python
def get_qnaContent(url):
    req = requests.get(url)
    cont = req.content
    soup = BeautifulSoup(cont, 'html.parser')
    
    qna = dict()
    qna['제목'] = soup.select_one('div.board_data.title_wrap > dl > dd').text
    qna['작성자'] = soup.select_one('dl.name > dd').text 
    qna['등록일'] =soup.select_one('dl.date > dd').text
    qna['조회수'] =soup.select_one('dl.count > dd').text
    qna['담당부서'] = soup.select_one('#conts > div.bbs_view > div:nth-of-type(3) > dl > dd').text.replace("담당부서","")
    qna['문의'] = soup.select_one('#conts > div.phone_box > p.phone').text.replace("전화번호","").replace(" : ", "")
    qna['답변'] = soup.select_one('#conts > div.bbs_view > div.bbs_article').text
  
    return qna
```

```python
def main():
    total_qnaList = list()
    fileName = '5_금융감독원(서민금융1332)'
    csv_file = open(str(fileName) + '.csv', "w", encoding = 'utf-8-sig', newline = '')
    
    columns = ['제목', '작성자', '등록일', '조회수', '담당부서', '문의', '답변']
    
    wr = csv.writer(csv_file)    
    wr.writerow(columns)
    
    for idx in tqdm_notebook(range(1, 5)):
        url = 'http://www.fss.or.kr/s1332/etc/list.jsp?sort=&orderby=&url=/s1332/1290645999591&bbsid=1290645999591&path2010=/fss/s1332&page=%s'%(str(idx))
        qnaList = get_qnaList(url)
        
        for qna in qnaList:
            qnaContent = get_qnaContent(qna)
            print(qnaContent)
            total_qnaList.append(qnaContent)
        print(idx, '페이지 완료')
    
    
    # '모든 수집 복지서비스'에 대한 반복문 -> 복지 서비스 하나
    for info_dict in total_qnaList:
        # 'columns'을 돌면서 info_dict가 해당 col을 가지고 있으면 정보를 적고, 없으면 NULL
        wr.writerow([info_dict.get(col, 'NULL') for col in columns])
    
    print('==================== 저장 완료 =======================')
    csv_file.close()
             
    
```

```python
if __name__ == "__main__":
    main()
```
