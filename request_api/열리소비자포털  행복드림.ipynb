## 페이지 번호 1 ~ 262 p

### 수집 칼럼
1. infoid(str): 게시물id
2. cntntsid(str): 메뉴ID
3. infodcsumry(str): 게시물에 대한 요약
4. infodetailcn
5. infoorigin(str): 출처
6. infosj(str): 
7. infourl(str): url

```python
f = open('output.csv', 'w', encoding='euc-kr', newline='')
wr = csv.writer(f)
wr.writerow(['게시물 ID', '메뉴 ID', '게시글 요약', 'detail', '출처', 'infosj', 'url'])
# wr.writeheader()

try:
    url = "https://www.consumer.go.kr/openapi/contents/index.do?serviceKey=인증키(XXXXXXXXXX)&pageNo=1&cntPerPage=10&cntntsId=0101"
    for page in tqdm_notebook(range(1, 263)):
        paramDict = { "serviceKey" : "인증키(XXXXXXXXXX)", "pageNo" : page, 'cntPerPage':100}
        response = requests.get(url, params=paramDict)
        print()
        print(page, "쪽", "status code :", response.status_code)
        print('================================================')
        
        r =response.text
        soup = BeautifulSoup(r, 'html.parser')
        
        # 하나의 contents에 10개 아이템 등록되어 있음.
        contents = soup.find_all('content')
        
        for cnt in contents:
            content_list = list()
            infoid = cnt.infoid.text
            try:
                cntntsid = cnt.cntntsid.text
            except:
                cntntsid = "NULL"
            try:
                infodcsumry = cnt.infodcsumry.text
            except:
                infodcsumry = "NULL"
            try:
                infodetailcn = cnt.infodetailcn.text
            except:
                infodetailcn = "NULL"
            try:
                infoorigin = cnt.infoorigin.text
            except:
                infoorigin = "NULL"
            try:
                infosj = cnt.infosj.text
            except:
                infosj = "NULL"
            try:
                infourl = cnt.infourl.text
            except:
                infourl = "NULL"
            content_list.append(infoid)
            content_list.append(cntntsid)
            content_list.append(infodcsumry)
            content_list.append(infodetailcn)
            content_list.append(infoorigin)
            content_list.append(infosj)
            content_list.append(infourl)

            wr.writerow(content_list)
            
    
except:
    print('page 정보 잘못됨')
    
f.close()  
```
